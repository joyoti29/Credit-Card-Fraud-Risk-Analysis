DATA CLEANING & EDA

```python
# import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# load the Data
df = pd.read_csv('E:/Project1/Credit Card Risk/Credit Card Fraud Risk Analysis.csv')

print("Number of Rows:",df.shape[0])
print("Number of Columns:",df.shape[1])
print(df.head())
print(df.info())
#check the missing value
print(df.isnull().sum())
#check the duplicate value
print(df.duplicated().sum())
#summary statistics of Numerical value
df.describe()
```

```python
# Clean column Names
df.columns = (
    df.columns
    .str.strip()
    .str.lower()
    .str.replace(" ", "_")
    .str.replace("(", "", regex=False)
    .str.replace(")", "", regex=False)
)

print(df.columns)
df
```

```python
#check data types
print(df.dtypes)
```

```python
# Convert Date Columns
df["transaction_date"] = pd.to_datetime(df["transaction_date"])
df
print(df.dtypes)
```

```python
#create date features
df["year"] = df["transaction_date"].dt.year
df["month_no"] = df["transaction_date"].dt.month
df["month_name"] = df["transaction_date"].dt.month_name()
df["quarter"] = df["transaction_date"].dt.quarter
df["day_name"] = df["transaction_date"].dt.day_name()
df["month_year"] = df["transaction_date"].dt.to_period("M").astype(str)
df
```

```python
#clean fraud type
df["fraud_type_clean"] = np.where(
    df["isfraud"] == 1,
    df["fraud_type"],
    "Not Fraud"
)
df
```

```python
#create amount bucket
df["amount_bucket"] = pd.cut(
    df["transaction_amount_inr"],
    bins=[0, 5000, 10000, 15000, 20000, 25000],
    labels=["0-5K", "5K-10K", "10K-15K", "15K-20K", "20K-25K"]
)
df
```

```python
#create fraud score bucket
df["fraud_score_bucket"] = pd.cut(
    df["fraud_score"],
    bins=[0, 25, 50, 75, 100],
    labels=["Low Score", "Medium Score", "High Score", "Very High Score"]
)
df
```

```python
#Risk rank
risk_map = {
    "Low": 1,
    "Medium": 2,
    "High": 3,
    "Critical": 4
}

df["risk_rank"] = df["fraud_risk"].map(risk_map)
df
```

```python
#KPI Calculation
total_transactions = df["transaction_id"].count()
fraud_transactions = df["isfraud"].sum()
fraud_rate = fraud_transactions / total_transactions
total_amount = df["transaction_amount_inr"].sum()
fraud_amount = df.loc[df["isfraud"] == 1, "transaction_amount_inr"].sum()

print("Total Transactions:", total_transactions)
print("Fraud Transactions:", fraud_transactions)
print("Fraud Rate:", round(fraud_rate * 100, 2), "%")
print("Total Amount:", total_amount)
print("Fraud Amount:", fraud_amount)
```

EDA ANALYSIS

```python
#Fraud by Category
category_fraud = df.groupby("transaction_category").agg(
    total_transactions=("transaction_id", "count"),
    fraud_transactions=("isfraud", "sum"),
    total_amount=("transaction_amount_inr", "sum")
)

category_fraud["fraud_rate"] = (
    category_fraud["fraud_transactions"] / category_fraud["total_transactions"] * 100
)

category_fraud.sort_values("fraud_rate", ascending=False)
```

```python
#Fraud by State
state_fraud = df.groupby("state").agg(
    total_transactions=("transaction_id", "count"),
    fraud_transactions=("isfraud", "sum")
)

state_fraud["fraud_rate"] = (
    state_fraud["fraud_transactions"] / state_fraud["total_transactions"] * 100
)

state_fraud.sort_values("fraud_rate", ascending=False)
```

```python
#Fraud by Card Type
card_fraud = df.groupby("card_type").agg(
    total_transactions=("transaction_id", "count"),
    fraud_transactions=("isfraud", "sum")
)

card_fraud["fraud_rate"] = (
    card_fraud["fraud_transactions"] / card_fraud["total_transactions"] * 100
)

card_fraud.sort_values("fraud_rate", ascending=False)
```

```python
#saved the clean file
df.to_csv("clean_credit_card_fraud_analysis.csv", index=False)
```
